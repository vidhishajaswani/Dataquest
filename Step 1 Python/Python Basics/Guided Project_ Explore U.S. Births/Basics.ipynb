```python
def read_csv(input_file):
    f=open(input_file)
    data=f.read()
    split_data=data.split('\n')
    string_list=split_data[1:]
    final_list=[]
    for each in string_list:
        int_fields=[]
        string_fields=each.split(',')
        for line in string_fields:
            int_fields.append(int(line))
        final_list.append(int_fields)
    return final_list

cdc_list=read_csv('US_births_1994-2003_CDC_NCHS.csv')
print(cdc_list[0:10])
```

```python
def month_births(list_of_list):
    birth_per_month={}
    for each in list_of_list:
        month=each[1]
        births=each[4]
        if month in birth_per_month:
            birth_per_month[month]=birth_per_month[month]+births
        else:
            birth_per_month[month]=births
    return birth_per_month
cdc_month_births=month_births(cdc_list)
cdc_month_births
```

```python
def dow_births(list_of_list):
    day_of_week={}
    for each in list_of_list:
        day=each[3]
        births=each[4]
        if day in day_of_week:
            day_of_week[day]=day_of_week[day]+births
        else:
            day_of_week[day]=births
    return day_of_week
cdc_day_births=dow_births(cdc_list)
cdc_day_births
```

```python
def calc_counts(data,column):
    dictionary={}
    for each in data:
        val=each[column]
        births=each[4]
        if val in dictionary:
            dictionary[val]=dictionary[val]+births
        else:
            dictionary[val]=births
    return dictionary
    
cdc_year_births=calc_counts(cdc_list,0)
cdc_month_births=calc_counts(cdc_list,1)
cdc_dom_births=calc_counts(cdc_list,2)
cdc_dow_births=calc_counts(cdc_list,3)
cdc_year_births
```
